```python
import json
import matplotlib.pyplot as plt

# Load data from epoch_log.txt
with open('./outputs/epoch_log.txt', 'r') as file:
    lines = file.readlines()

# Extract data from JSON format
data = [json.loads(line) for line in lines]

# Extract individual parameters for plotting
epochs = [entry['epoch'] for entry in data]
avg_d_loss = [entry['avg_d_loss'] for entry in data]
avg_g_loss = [entry['avg_g_loss'] for entry in data]
avg_dx = [entry['avg_dx'] for entry in data]
avg_dgx = [entry['avg_dgx'] for entry in data]

# Plot the graphs
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(epochs, avg_d_loss, label='Average D Loss')
plt.title('Average Discriminator Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(epochs, avg_g_loss, label='Average G Loss')
plt.title('Average Generator Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(epochs, avg_dx, label='Average D(X)')
plt.title('Average D(X)')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(epochs, avg_dgx, label='Average D(G(X))')
plt.title('Average D(G(X))')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend()

plt.tight_layout()
plt.show()
```
